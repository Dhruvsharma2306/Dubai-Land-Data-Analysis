```python
import geopandas as gpd
import folium

# Load both KML files
gdf_community = gpd.read_file("C:/Users/DSHA-EXT/Desktop/Dubai Land Data/GIS/Community.kml", driver='KML')
gdf_sector = gpd.read_file("C:/Users/DSHA-EXT/Desktop/Dubai Land Data/GIS/Sectors.kml", driver='KML')

# Create the folium map centered between both layers
center = [
    (gdf_community.geometry.centroid.y.mean() + gdf_sector.geometry.centroid.y.mean()) / 2,
    (gdf_community.geometry.centroid.x.mean() + gdf_sector.geometry.centroid.x.mean()) / 2,
]
m = folium.Map(location=center, zoom_start=12)

# Add Community layer
for _, row in gdf_community.iterrows():
    folium.GeoJson(
        row['geometry'],
        name='Communities',
        style_function=lambda x: {'fillColor': 'blue', 'color': 'blue', 'weight': 1, 'fillOpacity': 0.3},
        tooltip=row.get('Name', 'No name'),
        popup=folium.Popup(str(row), max_width=300),
    ).add_to(m)

# Add Sector layer
for _, row in gdf_sector.iterrows():
    folium.GeoJson(
        row['geometry'],
        name='Sectors',
        style_function=lambda x: {'fillColor': 'green', 'color': 'green', 'weight': 1.5, 'fillOpacity': 0.2},
        tooltip=row.get('Name', 'No name'),
        popup=folium.Popup(str(row), max_width=300),
    ).add_to(m)

# Add layer control so user can toggle
folium.LayerControl().add_to(m)

# Save and view the map
m.save("dubai_communities_sectors_map.html")
```

```python
import geopandas as gpd
import folium

# --- Load and Filter ---
gdf_community = gpd.read_file("C:/Users/DSHA-EXT/Desktop/Dubai Land Data/GIS/Community.kml", driver='KML')
gdf_sector = gpd.read_file("C:/Users/DSHA-EXT/Desktop/Dubai Land Data/GIS/Sectors.kml", driver='KML')

# Filter examples
gdf_community = gdf_community[gdf_community['Name'].str.contains("Jumeirah", na=False)]
gdf_sector = gdf_sector[gdf_sector['Name'].isin(["Sector 1", "Sector 2"])]

# --- Map setup ---
center = [
    (gdf_community.geometry.centroid.y.mean() + gdf_sector.geometry.centroid.y.mean()) / 2,
    (gdf_community.geometry.centroid.x.mean() + gdf_sector.geometry.centroid.x.mean()) / 2,
]
m = folium.Map(location=center, zoom_start=13, tiles=None)

# --- Satellite Layer ---
folium.TileLayer('Esri.WorldImagery', name='Satellite').add_to(m)

# --- Community Layer ---
for _, row in gdf_community.iterrows():
    folium.GeoJson(
        row['geometry'],
        name='Communities',
        style_function=lambda x: {'fillColor': 'blue', 'color': 'blue', 'weight': 1, 'fillOpacity': 0.3},
        tooltip=row.get('Name', 'No name'),
        popup=folium.Popup(str(row), max_width=300),
    ).add_to(m)

# --- Sector Layer ---
for _, row in gdf_sector.iterrows():
    folium.GeoJson(
        row['geometry'],
        name='Sectors',
        style_function=lambda x: {'fillColor': 'green', 'color': 'green', 'weight': 1.5, 'fillOpacity': 0.2},
        tooltip=row.get('Name', 'No name'),
        popup=folium.Popup(str(row), max_width=300),
    ).add_to(m)

# --- Add Layer Control ---
folium.LayerControl().add_to(m)

# --- Add Legend ---
legend_html = '''
<div style="position: fixed; 
            bottom: 50px; left: 50px; width: 150px; height: 90px; 
            background-color: white; border:2px solid grey; z-index:9999; font-size:14px;
            padding: 10px;">
  <b>Legend</b><br>
  <i style="background:blue;opacity:0.7;width:10px;height:10px;display:inline-block;"></i> Communities<br>
  <i style="background:green;opacity:0.7;width:10px;height:10px;display:inline-block;"></i> Sectors
</div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

# --- Save ---
m.save("dubai_filtered_map.html")
```
