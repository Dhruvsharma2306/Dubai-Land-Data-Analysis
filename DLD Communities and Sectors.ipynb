```python
import geopandas as gpd

# Load the shapefile
gdf_communities = gpd.read_file("C:/Users/DSHA-EXT/Desktop/Dubai Land Data/GIS/Community.kml")

# Display the first few rows
gdf_communities.head(100)
```

```python
gdf_communities['Description'][0]
```

```python
# Ensure community shapefile is also in EPSG:4326 (WGS84)
gdf_communities = gdf_communities.to_crs('EPSG:4326')
```

```python
import re
```

```python
# Define a function to extract community number using regex
def extract_community_number(description):
    match = re.search(r'<td>COMM_NUM</td>\s*<td>(\d+)</td>', description)
    if match:
        return match.group(1)
    return None

# Apply the function to create a new column 'community_number'
gdf_communities['community_number'] = gdf_communities['Description'].apply(extract_community_number)
```

```python
import geopandas as gpd

# Make sure both GeoDataFrames are in the same CRS
gdf_communities = gdf_communities.to_crs('EPSG:4326')
gdf_sectors = gdf_sectors.to_crs('EPSG:4326')

# Spatial join: assign sector to each community
gdf_communities_with_sector = gpd.sjoin(
    gdf_communities,
    gdf_sectors[['Name', 'geometry']],  # use only necessary columns
    how='left',
    predicate='within'
)

# Rename the joined sector column for clarity
gdf_communities_with_sector.rename(columns={'Name_left': 'Community_Name', 'Name_right': 'Sector_Name'}, inplace=True)

# Optional: preview the result
gdf_communities_with_sector.head(224)
```

```python
gdf_communities_with_sector = gdf_communities_with_sector[['Community_Name','Sector_Name','geometry','community_number']]
```

```python
# Spatial join
gdf_joined_2 = gpd.sjoin(gdf_areas, gdf_communities_with_sector, how='left', predicate='within')

# Check the result
gdf_joined_2[['area_name_en', 'Community_Name','community_number','Sector_Name']].head(200)
```
